## Tutorial for DaisyRec-v2.0
- This is a basic tutorial for how to use DasiyRec-v2.0
- It includes two parts, i.e., hyper-parameter tuning and test
- Let's take __NeuMF__ runing on __ML-1M__ as an example

### Hyper-parameter Tuning
- ___Step 1___: Open [tune_command](http://daisyrecguicommandgenerator.pythonanywhere.com/tune_command) in the GUI Command Generator
- ___Step 2___: Fill in 'Basic Settings' and click the 'Submit' Button

- ___Step 3___: Tick 'Algorithm-Specific Settings' and click 'Submit' Button


- ___Step 4___: Set the hyper-parameter searching space and click 'Generate' Button
    - For int/float, simply set \['Min. Value', 'Max. Value'\]
    - For int/float choice, set \['Min. Value', 'Max. Value', 'step size'\] or directly set \['choice set'\]



- ___Step 5___: Paste the generated command into the terminal to start tune parameter with HyperOpt


- ___Step 6___: After tuning, the results are automatically saved under './tune_log/CL_neumf_ml-1m_5filter_tloo.csv'
- ___Step 7___: Results on the validation set of 30 rounds are available, and select the parameter settings with the best NDCG

### Test Procedure
- ___Step 1___: Open [test_command](http://daisyrecguicommandgenerator.pythonanywhere.com/test_command) in the GUI Command Generator
- ___Step 2___: Fill in 'Basic Settings' (same as tune) and click 'Submit' Button
- ___Step 3___: Fill in 'Algorithm-Specific Settings' with the best parameter settings and click 'Generate' Button


- ___Step 4___: Paste the generated test command into the terminal to start test


- ___Step 5___: After testing, the results are saved under './res/ml-1m/5-filter/tloo/CL_point_neumf.csv'
